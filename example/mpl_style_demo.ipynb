```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
x = np.linspace(0, 2. * np.pi)
ys = [np.sin(x + phi * np.pi) for phi in [1. / 4., 1. / 2., 3. / 4.]]
```

```python
def make_plot(x, ys):
    fig, ax = plt.subplots(1, 1)
    for y in ys:
        ax.plot(x, y) 
    return fig, ax
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use('blt_paper')
fig, ax = make_plot(x, ys)
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use('blt_pres')
fig, ax = make_plot(x, ys)
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use('idl')
fig, ax = make_plot(x, ys)
ax.set_title(r"$f(x) = \sin(x)$, attempt at idl")

fig, ax = make_plot(x, ys)
ax.set_title(r"$f(x) = \sin(x)$, attempt at idl")
ax.set_xscale('log')
plt.show()
```
