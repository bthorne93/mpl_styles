```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2. * np.pi)
ys = [np.sin(x + phi * np.pi) for phi in [1. / 4., 1. / 2., 3. / 4.]]
```

```python
def make_plot(x, ys):
    fig, ax = plt.subplots(1, 1)
    for y in ys:
        ax.plot(x, y) 
    return fig, ax
```

```python
plt.style.use('blt')
fig, ax = make_plot(x, ys)
```

```python
plt.style.use('blt_paper')
fig, ax = make_plot(x, ys)
```

```python
plt.style.use('blt_pres')
fig, ax = make_plot(x, ys)
```

```python
plt.style.use('idl')
fig, ax = make_plot(x, ys)
```
